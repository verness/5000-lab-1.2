## Key concept
- **Machine Learning**: The study of algorithms and statistical models that enable computers to perform tasks without explicit instructions, relying on patterns and inference instead.
- **Data Science**: An interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from structured and unstructured data.
- **Artificial Intelligence (AI)**: The simulation of human intelligence processes by machines, especially computer systems, which includes learning, reasoning, and self-correction.

## Data Leakage

Data leakage happens when information from outside the training fold sneaks into model training, making performance look unrealistically good. Common causes include scaling or imputation fit on the full dataset, target-encoded features computed with all rows, or time-ordered data shuffled incorrectly.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(3*x)

plt.figure()
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
